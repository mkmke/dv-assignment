```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
np.random.seed(0)
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value_1': np.random.randint(1, 100, 5),
    'Value_2': np.random.randint(1, 100, 5),
    'Value_3': np.random.randint(1, 100, 5)
}
df = pd.DataFrame(data)
df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Category'], df['Value_1'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Value_1'], bins=10, color='salmon', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Value_1'], df['Value_2'], s=100, c='purple', alpha=0.5)
plt.xlabel('Value 1')
plt.ylabel('Value 2')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop('Category', axis=1))
plt.title('Box Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Category'], df['Value_1'], marker='o', linestyle='-', color='green')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Line Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.drop('Category', axis=1), cmap='coolwarm', annot=True)
plt.title('Heatmap')
plt.show()
```

```python
labels = df.columns[1:]
stats = df.iloc[0, 1:].values
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
stats = np.concatenate((stats,[stats[0]]))
angles += angles[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, stats, color='red', alpha=0.25)
ax.plot(angles, stats, color='red', linewidth=2)  # The outline
ax.set_yticklabels([])
plt.title('Radar Map')
plt.show()
```
